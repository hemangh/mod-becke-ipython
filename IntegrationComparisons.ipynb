```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
directory = "/home/gharibnejad/Documents/DataSets/LucaPaper/"
data2 = pd.read_csv(directory+"sinTestBecke.csv",delimiter=',',dtype={'float': 'str'},skiprows=[0])
data2
```

```python
f = plt.figure(figsize=(18, 15))
ax = f.add_subplot()
ax.tick_params(axis="x", labelsize=18)
ax.tick_params(axis="y", labelsize=18)
label1 = "% diff"
label2 = "abs. diff"
ax.set_ylabel('Error', labelpad = 20,fontsize=24)
ax.set_xlabel('Num. angular pts', labelpad = 20,fontsize=24)
ax.loglog(data1["Angular #pts"],data1["Sin %diff err"],'bo-')
ax.loglog(data1["Angular #pts"],data1["Sin abs diff"],'ko--')
ax.legend([label1, label2],title="Sin(k r) at one centeral grid",title_fontsize=30,fontsize=20, fancybox=True)
```
