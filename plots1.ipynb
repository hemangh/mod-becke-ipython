```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from pandas import DataFrame
import matplotlib.gridspec as gridspec
import math as mt
os.chdir('/Users/hng5/ProgramDev/gridcheck/build')
```

```python
fig = plt.figure(figsize=(30, 20))
gs = gridspec.GridSpec(3, 4,
                       width_ratios=[1,1,1,1.2],hspace=0.5
                       )

w, h = 4, 3;
dat = [[0 for x in range(w)] for y in range(h)] 
k = 0
for i in range(1,4):

    for j in range(1,5):

        #dat[i-1][j-1]='dat'+str(i)+str(j)
        dat[i-1][j-1]=pd.read_csv('./Alpha_lmax_exp_'+str(j)+'R_'+str(i)+'.0.txt', delimiter = ',')
        dat[i-1][j-1].set_index('    lmax',inplace=True)
        #print(dat[i-1][j-1].head())
        ax = plt.subplot(gs[k])
        k = k + 1
        if(j == 4):
            pcm = ax.pcolor(dat[i-1][j-1],
                   norm=colors.LogNorm(vmin=1E-13, vmax=1.E-1),
                   cmap='PuBu_r')
            plt.colorbar(pcm,extend='max')
        else:    
            ax.pcolor(dat[i-1][j-1],
                   norm=colors.LogNorm(vmin=1E-13, vmax=1.E-1),
                   cmap='PuBu_r')
        ax.set_title('$R$='+str(i)+'  $n$='+str(j),fontsize=24)
        ax.set_xticks(np.arange(0.5, len(dat[i-1][j-1].columns), 1))
        ax.set_xticklabels(dat[i-1][j-1].columns,rotation=90,fontsize=16)
        ax.set_yticks(np.arange(0.5, len(dat[i-1][j-1].index), 1))
        ax.set_yticklabels(dat[i-1][j-1].index[0:],fontsize=16)
        ax.set_xlabel(r'$\alpha$',fontsize=20)
        ax.set_ylabel("lmax",fontsize=20)
        
            
plt.savefig('Fig-powers1to4.pdf', bbox_inches='tight')
#print(dat)
        
```
