```python
import math as mt
from scipy.interpolate import make_interp_spline, BSpline
def fitting_func(ir,n,Rmax):
    ans=[]
    for x in ir:
        ans.append( mt.exp(mt.log(1E-20)*mt.pow((x/Rmax),(2*n))))
    return ans

fig, ax1 = plt.subplots()
#plt.tight_layout()
#plt.style.use('ggplot')
rmax = 1
x= np.arange(0.,1 , 0.01)
#print(x)
y = fitting_func(x,1,1)
#print(y)
# for n in range(1,5):
#     y.append( fitting_func(n,rmax))


ax1.set_ylim([1.E-20,1])
ax1.set_xlim([0.,1.0])
ax1.plot(x,y,  c='#F5793A',linewidth=3)
ax1.text(0.12,0.4,'n=2',fontsize=15,rotation=-70)


y = fitting_func(x,2,1)
ax1.plot(x,y,  c='#A95AA1',linewidth=3)
ax1.text(0.35,0.4,'n=4',fontsize=15,rotation=-70)

y = fitting_func(x,3,1)
ax1.plot(x,y,  c='#85C0F9',linewidth=3)
ax1.text(0.50,0.4,'n=6',fontsize=14,rotation=-70)
# y = fitting_func(x,4,1)
# ax1.plot(x,y,  c='#0F2080',linewidth=3)
# ax1.text(0.59,0.4,'n=8',fontsize=14,rotation=-70)
y = fitting_func(x,5,1)
ax1.plot(x,y,  c='k',linewidth=3)
ax1.text(0.66,0.35,'n=10',fontsize=14,rotation=-70)

# ax1.plot(x,y[2], c='g', marker="v", ms=10, label='n=3')
# ax1.plot(x,y[3],  c='orange', marker="^", ms=10, label='n=4')
# ax1.plot(x,y[4],  c='black', marker="p", ms=10, label='n=5')
plt.legend(loc='best')
ax1.yaxis.major.formatter._useMathText = True
#ax1.set_yscale('log')
ax1.set_xlabel('$r$',fontsize=20)
ax1.set_ylabel("switching function", fontsize=20)
plt.xticks()
plt.yticks(fontsize=14)
#plt.annotate(r"R$_{At}$",[0.91,0.1],fontsize=20)
# plt.show()
ax1.set_xticklabels([' ',' ',' ',' ',' ',r"R$_{At}$"],fontsize=20)

ax1.get_legend().remove()
#ax1.set_xticklabels([r"R$_{At}$"])
plt.savefig('Fig-powers.pdf', bbox_inches='tight')
```
